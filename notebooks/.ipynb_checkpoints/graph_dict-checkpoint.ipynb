# Create Graph Dictionary

## Single path DFS

```python
gd = {}
gd_prop = {}
ht = 50
wd = 50
x_list = []
y_list = []
xy_list = []
for x in range(ht):
    for y in range(wd):
        loc = (y,x)
        x_list.append(x)
        y_list.append(y)
        xy_list.append(loc)
        surr = []
        for tls in [(-1,0),(0,-1),(1,0),(0,1)]:
            y2 = y + tls[0]
            x2 = x + tls[1]
            if y2 >= 0 and x2 >=0 and y2 < ht and x2 < wd:
                surr.append((y2,x2))
        gd[loc] = surr
        gd_prop[loc] = 0
```

```python
loc = [(np.random.choice(ht), np.random.choice(wd))]
loc_lines = []
while len(loc)< int(1*len(gd)):
    locp = 0
    free_loc = []
    while len(free_loc) == 0:
        locp -= 1
        free_loc = [x for x in gd[loc[locp]] if x not in loc]
    new_loc = free_loc[np.random.choice(len(free_loc))]
    loc_lines.append(tuple(x/2 for x in tuple(map(op.add, loc[locp], new_loc))))
#     loc_lines.append( (loc[locp], new_loc))
    loc.append(new_loc)
```

```python
xx = []
yy = []
for x in np.arange(-.5,wd,0.5):
    for y in np.arange(-.5,ht,0.5):
        xx.append(x)
        yy.append(y)
df_lines = pd.DataFrame({'x':xx,'y':yy })
df_lines['col'] = 0
df_lines['yx'] = list(zip(df_lines.y,df_lines.x))
df_lines.loc[df_lines.yx.isin(loc),'col'] = 1
df_lines.loc[df_lines.yx.isin(loc_lines),'col'] = 1
df_lines.loc[df_lines.yx.isin(loc[:5]),'col'] = 2
df_lines.loc[df_lines.yx==loc[500],'col'] = 2
```

```python
ggplot(df_lines, aes(x='x', y='y', fill='col')) + geom_tile(show_legend=False) + blank_theme
```

## Multi Path DFS

```python
gd = {}
gd_prop = {}
ht = 50
wd = 50
x_list = []
y_list = []
xy_list = []
for x in range(ht):
    for y in range(wd):
        loc = (y,x)
        x_list.append(x)
        y_list.append(y)
        xy_list.append(loc)
        surr = []
        for tls in [(-1,0),(0,-1),(1,0),(0,1)]:
            y2 = y + tls[0]
            x2 = x + tls[1]
            if y2 >= 0 and x2 >=0 and y2 < ht and x2 < wd:
                surr.append((y2,x2))
        gd[loc] = surr
        gd_prop[loc] = 0
```

```python
loc_d = {}
loc_lines_d = {}
line_ct = 50
loc = list(zip(np.random.choice(ht,line_ct,replace=False),np.random.choice(ht,line_ct,replace=False)))
for i in range(line_ct):
    loc_d[i] = [loc[i]]
    loc_lines_d[i] = []
# loc_lines = []
dni = []
while len(loc) < int(0.9*len(gd)):
    for i in range(line_ct):
        if i not in dni:
            locp = 0
            free_loc = []
            while len(free_loc) == 0 and locp*-1 < len(loc_d[i]):
                locp -= 1
                free_loc = [x for x in gd[loc_d[i][locp]] if x not in loc]
            if len(free_loc) > 0:
                new_loc = free_loc[np.random.choice(len(free_loc))]
                loc_lines_d[i].append(tuple(x/2 for x in tuple(map(op.add, loc_d[i][locp], new_loc))))
                loc_d[i].append(new_loc)
                loc.append(new_loc)
            else:
                dni.append(i)
# [len(loc_d[i]) for i in range(line_ct)]
```

```python
xx = []
yy = []
for x in np.arange(-.5,wd,0.5):
    for y in np.arange(-.5,ht,0.5):
        xx.append(x)
        yy.append(y)
df_lines = pd.DataFrame({'x':xx,'y':yy })
df_lines['col'] = -5
# df_lines['col2'] = 0
df_lines['yx'] = list(zip(df_lines.y,df_lines.x))
for i in range(line_ct):
    df_lines.loc[df_lines.yx.isin(loc_d[i]),'col'] = i
    df_lines.loc[df_lines.yx.isin(loc_lines_d[i]),'col'] = i
#     df_lines.loc[df_lines.yx.isin(loc_d[i]),'col2'] = 1
#     df_lines.loc[df_lines.yx.isin(loc_lines_d[i]),'col2'] = 1
# df_lines.loc[df_lines.yx.isin(loc[:5]),'col'] = 2
# df_lines.loc[df_lines.yx==loc[500],'col'] = 2
```

```python
ggplot(df_lines, aes(x='x', y='y', fill='col')) + geom_tile() + blank_theme 
```
